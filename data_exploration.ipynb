```python
%%time
X_train, X_test, features = get_data()
```

```python
from numpy import genfromtxt
pred_pre = genfromtxt('submissions/Cantab-17-input.predict', delimiter=',')
pred_post = genfromtxt('submissions/Cantab-19-input.predict', delimiter=',')
```

```python
X_test.insert(0, "pred_post", pred_post)
X_test.insert(0, "pred_pre", pred_pre)
```

```python
X_test[['Converted']] = X_test[['Converted']].astype(float)
```

```python
X_sorted = X_test.sort_values(by='user')
X_sorted['user'] = pd.factorize(X_sorted['user'])[0]
# X_sorted = X_sorted.set_index('user')
```

```python
quotes_converted = X_sorted.groupby('user').apply(lambda x: (len(x), np.sum(x['Converted'])))
```

```python
quotes_converted_arr =  np.zeros((len(quotes_converted), 2))
for i in range(len(quotes_converted)):
    quotes_converted_arr[i, 0] =  quotes_converted[i][0]
    quotes_converted_arr[i, 1] =  quotes_converted[i][1]
```

```python
plt.scatter(quotes_converted_arr[:,1], quotes_converted_arr[:,0])
```

```python
quotes_converted_pd = pd.DataFrame(quotes_converted_arr, columns=['NumberQuotesFromUser', 'UserConversions'])
```

```python
sns.jointplot(data=quotes_converted_pd,
              x='NumberQuotesFromUser', y='UserConversions');
plt.savefig('quotes_vs_conversations.pdf')
```

```python
sns.barplot(data=quotes_converted_pd, x='NumberQuotesFromUser', y='UserConversions')
```
